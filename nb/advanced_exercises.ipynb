````python
MENU = {'sandwich': 10, 'tea': 7, 'salad': 9}
 
def restaurant():
    total = 0
    while True:
        order = input('Order: ').strip()
 
        if not order:
            break
 
        if order in MENU:
            price = MENU[order]
            total += price
            print(f'{order} is {price}, total is {total}')
 
        else:
            print(f'We are fresh out of {order} today')
 
    print(f'Your total is {total}')

restaurant()
````

````python
def get_rainfall():
    rainfall = {}
 
    while True:
        city_name = input('Enter city name: ')
        if not city_name:
            break
 
        mm_rain = input('Enter mm rain: ')
        rainfall[city_name] = rainfall.get(city_name, 0) + int(mm_rain)
 
    for city, rain in rainfall.items():
        print(f'{city}: {rain}')

get_rainfall()
````

````python
def dictdiff(first, second):
    output = {}
    all_keys = first.keys() | second.keys()  # Vereinigung der beiden key-Mengen
 
    for key in all_keys:
        if first.get(key) != second.get(key):
            output[key] = [first.get(key), second.get(key)]
    return output
 

d1 = {'a':1, 'b':2, 'c':3}
d2 = {'a':1, 'b':2, 'c':4}
assert dictdiff(d1, d1) == {}  # {} no diff
assert dictdiff(d1, d2) == {"c": [3,4]}

d3 = {'a':1, 'b':2, 'd':3}
d4 = {'a':1, 'b':2, 'c':4}
assert dictdiff(d3, d4) == {"c": [None, 4], "d": [3, None]}

d5 = {'a':1, 'b':2, 'd':4}
assert dictdiff(d1, d5) == {"c": [3, None], "d": [None, 4]}
````

### Solutions

````python
def get_final_line(filename):
    final_line = ''
    for current_line in open(filename):
        final_line = current_line.strip()
    return final_line


assert get_final_line("../data/faust.txt") == "subscribe to our email newsletter to hear about new eBooks."
````
